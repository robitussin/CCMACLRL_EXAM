## Evaluation



Submissions are scored on the Root Mean Squared Error (RMSE).

RMSE is defined as:

$$RMSE = \Bigg( \frac{1}{N} \displaystyle\sum_{i=1}^{N} (y_i−\hat{y}_i)^2 \Bigg)^\frac{1}{2}$$

Where:

$\hat{y}_i$ is the predicted value

$y_i$ is the original value for each instance $i$.
